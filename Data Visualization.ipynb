# Data Visualization

### Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Bar Chart

```python
# create a dataframe
data_1 = {
    'Country':['USA','SGP','JPN','IDN'], 
    'GDP':[40000,35000,50000,25000]
}

df_1 = pd.DataFrame(data_1,columns=['Country','GDP'])
df_1
```

```python
# menggunakan pandas
df_1.plot(x='Country', y='GDP', kind='bar')
```

```python
df_1.plot(x='Country', y='GDP', kind='barh')
```

```python
# menggunakan matplotlib
plt.bar(df_1['Country'], df_1['GDP'],color='green')
plt.xlabel('Country')
plt.ylabel('GDP')
plt.title('GDP per Country')
plt.show()
```

```python
# menggunakan matplotlib
plt.bar(df_1['Country'], df_1['GDP'],color=['green', 'blue', 'red', 'yellow'])
plt.xlabel('Country', fontsize=12)
plt.ylabel('GDP', fontsize=12)
plt.title('GDP per Country', fontsize=18)
plt.show()
```

```python
sns.barplot(x=df_1['Country'], y=df_1['GDP'])
plt.title('GDP per Country')
plt.show()
```

```python
# create a dataframe
data_2 = {
    'Year':[1930,1920,1940,1950], 
    'GMV':[40000,35000,50000,25000]
}

df_2 = pd.DataFrame(data_2,columns=['Year','GMV'])
df_2
```

```python
df_2_sorted = df_2.sort_values(by='Year')
```

```python
df_2.plot(x='Year', y='GMV', kind='bar')
```

### Line Chart

```python
plt.plot(df_2_sorted['Year'], df_2_sorted['GMV'], color='green', marker='o')
plt.xlabel('Country', fontsize=12)
plt.ylabel('GDP', fontsize=12)
plt.title('GDP per Country', fontsize=18)
plt.show()
```

```python
sns.lineplot(x=df_2['Year'], y=df_2['GMV'])
plt.show()
```

### Pie Chart

```python
# create a dataframe
data_4 = {'Tasks': [300,500,700]}
df_4 = pd.DataFrame(data_4,columns=['Tasks'],index = ['Tasks Pending','Tasks Ongoing','Tasks Completed'])
df_4
```

```python
plt.pie(df_4['Tasks'], labels=['Task Pending', 'Task Ongoing', 'Task Completed'], autopct='%.2f%%')
plt.show()
```

### Using Dataset

```python
# Import Dataset
data = pd.read_csv('D:\Digital Skola\Dataset\Ritz_Jager_Data.csv') 
data.head(10)
```

```python
# Dari negara mana paling banyak pengunjung hotel berasal ?

# step 1 - filter check out
data_checkout = data[data['reservation_status'] == 'Check-Out']
data_checkout

# step 2 - aggregasi jumlah pengunjung per negara
data_country_agg = data_checkout.groupby("country_origin").agg(count = pd.NamedAgg(column='country_origin', aggfunc='count'))

# step 3 - sorting
data_country_agg_sorted = data_country_agg.sort_values(['count'], ascending=False)

# step 4 - ambil 10 besar
data_country_top10 = data_country_agg_sorted.head(10).reset_index()

# step 5 - visualisasi
plt.figure(figsize=(16,8))
plt.bar(data_country_top10['country_origin'], data_country_top10['count'])
plt.show()
```
